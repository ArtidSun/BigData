# k-Means

```python
def k_means(fm: str,k: int):
    data = pd.read_csv(fm,header=None)
    data.columns = ['x','y']
    
    model_kmeans = cluster.KMeans(n_clusters=k, max_iter=50, random_state=1)
    model_kmeans.fit(data)

    data['cluster_id'] = model_kmeans.labels_

    centroids = model_kmeans.cluster_centers_
    print(centroids)

    plt.scatter(data['x'],data['y'], c=data['cluster_id'])
    plt.scatter(centroids[:,0],centroids[:,1],marker='X', c='r')
    plt.show()
```

```python
k_means('data2Dset1.csv',3)
```

# Hierarchical Clustering

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import dendrogram, linkage, fcluster
from scipy.spatial import distance_matrix
import pandas as pd
```

```python
def h_clustering(fm: str,m: str,t: int):
    data = pd.read_csv(fm,header=None)
    data.columns = ['x','y']
    print(data.index)
    linkage_data = linkage(data, method=m, metric='euclidean')
    dendrogram(linkage_data,labels=data.index.to_list())

    plt.title('method='+m)
    plt.show()
    
    cluster_id = fcluster(linkage_data, t=t, criterion='distance')
    plt.scatter(data['x'],data['y'], c=cluster_id)
    plt.show()
```

```python
h_clustering('data2Dset1.csv','single',1.5)
```

```python
h_clustering('data2Dset1.csv','average',3)
```

```python
h_clustering('data2Dset1.csv','complete',5)
```

```python
h_clustering('data2Dset2.csv','single',0.16)
```
